### Duplicates

```python
## function to find duplicates

def find_duplicates(df, subset=None):
    """
    Identifies rows in a DataFrame where all data in the specified columns are identical.
    """
    duplicates = df[df.duplicated(subset=subset, keep=False)]
    if duplicates.empty:
        print(f"No duplicate found based on selected columns.")
        return pd.DataFrame()  
    print(f"Found {len(duplicates)} duplicate found based on selected columns")
    
    return duplicates

exact_duplicates = find_duplicates(df, subset=None)

exact_duplicates
```

```python
# There are now 67 categories hence this will be rescaled to 9 categories
def classify_wptype(x):
    if (x == 'Handpump'or x == 'handpump'or x == 'hand pump'or x == 'protec dug_well'):
        return 'protected dug_well'
    elif (x =='Public public public tap/standpipe (standalone or water kiosk'
          or x == 'Piped water into dwelling/plot/yard'or x == 'public public public tap'
          or x == 'public public tap'or x == 'public public public tap'):
        return 'public tap/standpipe'
    elif (x == 'unprotec dug_well'):
        return 'Unprotected dug_well'
    elif (x == 'borewell'or x == 'Tube well or borehole'):
          return 'borewell'
    elif (x == 'Sand/Subsurface gravity_scheme (with well or standpipe)'
          or x == 'gravity_scheme'):
          return 'gravity scheme'
    elif (x == 'spring_protec'):
          return 'protected spring'
    elif (x == 'Rainwater (harvesting)' or x == 'Dam/pan (runoff harvesting)'):
          return 'rainwater harvesting'
    elif (x == 'water tank'):
          return 'water tank'
    else:
        return 'OTHER'
    
gdf1['wp_type'] = gdf1['wp_type'].apply(classify_wptype)
print(gdf1['wp_type'].value_counts())
```

### maintenance 

```python
# rescaling the maintenance features. 199 categories will be rescaled to 16 categories
def classify_maintainance(x):
    if (x == 'Community' or x == 'community people' or x == 'Community itself'
        or x == 'COMMUNITY' or x == 'community' or x == 'Community People'
        or x == 'Community leaders'or x == 'Community leaders'):
        return 'Community'
    elif (x == 'Privat owner Other' or x == 'WASH management committee'
          or x == 'School' or x == 'No management' or x == 'NGO' or x == 'Health Facility'
          or x == 'Uknown' or x == 'SALWACO' or x == 'Other Institution'
          or x == 'Local council' or x == 'GUMA'):
        return x 
    else:
        return 'other'
gdf1['maintenance'] = gdf1['maintenance'].apply(classify_maintainance)
print(gdf1['maintenance'].value_counts())
```

### extract_type

```python
print(gdf1['extract_type'].value_counts().to_dict())
```

```python
# rescaling the extract_type from 176 categories

## hydram pump = hydraulic ram pump, 
# water trucking = water truck delivering water, 
def classify_extra_type(x):
    if (x == 'Hand pump' or x == 'Gravity' 
       or x == 'Hydram pump' or x == 'Submersible pump'
       or x == 'Surface pump'):
        return x 
    elif (x == 'Hand manual (e.g. rope pump, rope & bucket)'):
        return 'Hand manual (e.g. rope pump, rope & bucket)'
    elif (x == 'tap'or x == 'run through tap'or x == 'RUN IN PIPE'or 
          x == 'RUNS THROUGH TAP' or x == 'RUNNING TAP'or x == 'Tap'
          or x == 'RUN THROUGH PIPE'):
        return 'Tap'
    elif (x == 'supplied by bowser' or x == 'water trucking'):
        return 'Water Trucking'
    else:
        return 'OTHER'
gdf1['extract_type'] = gdf1['extract_type'].apply(classify_extra_type)
print(gdf1['extract_type'].value_counts())
```

### wp_billed

```python
print(gdf1['wp_billed'].value_counts().to_dict())
```

```python
# rescaling the maintenance features. There are 199 categories hence this will be rescaled to 16 categories

def classify_wp_billed(x):
    if (x == 'No payment – its free'):
        return 'No payment' 
    elif (x == '<Null>'):
        return 'Unknown'
    elif (x == 'Yes, Flat rate (LE/HH/month)'):
        return 'Yes, Monthly'
    elif (x == 'Yes, Per jerrican 20L' or x == 'Yes, Metered (LE/m3)'):
        return x
    else:
        return 'OTHER'
    
gdf1['wp_billed'] = gdf1['wp_billed'].apply(classify_wp_billed)
# df2.head()
print(gdf1['wp_billed'].value_counts())
```

### const_year

Most of the wells were constructed in the 2000's. 
After 2009, construction per year exceeds 1000. Therefore, no further classification was done after 2010

```python
## convert year from string to numerical
# df2['const_year'] = df2['const_year'].apply(pd.to_numeric, errors='coerce', axis=1)
gdf1.const_year = gdf1.const_year.astype(int)
print(gdf1['const_year'].value_counts())
```

```python
# rescaling the const_year features from 76 categories to n
def classify_const_year(x):
        if (x < 1950):
            return 'Before 1950'
        elif (x < 2000):
            return '1950 - 2000'
        elif (x < 2005):
            return '2001 - 2005'
        elif (x < 2010):
            return '2005 - 2010'
        else:
            return x
        
gdf1['const_yr'] = gdf1['const_year'].apply(classify_const_year)
print(gdf1['const_yr'].value_counts())
```

### Summary

**Data structure**

The raw survey data contains 52 columns of WASH and ODF information. The data is formatted so that each row corresponds to an individual respondent and a column for each question. Columns are in the standard form of *code|Field name* for example *7430032|Water point Name*. Data in these columns are typically coded responses

*Reformatting column names was done*

Columns that contains coded responses are categorical. For example in the column *3830002|Observe what device is present for hand washing the data* responses was  recorded as `1:No device present`

*Seperate coded responses in categorical columns was done*

**Data type**

There are 4 numerical data and 48 categorical data, based on data format. The numerical data columns represent Location information (Longitude and Latitude), Construction date of the water point, and average time for repairs for a broken water point.

*Construction year was converted to year and intervals*

**Missing values**

There are only 3 columns with missing values: 

   - *1410054|Number of taps at this point:* 0.01% missing values
   - *6430041|Others Installer / implementing agency:* 1.34% missing values
   - *7430040|Is there a trained mechanic available at this point?:* 0.06% missing values 
        
**Categories in categorical features**

Categorical columns without coded responses contains varied amount of categories. For example the `Type of water point`column has 94 unique values but it is unlikely that there are 94 different types of water points implemented in the country.

*Merging of these categories was done based on my experience, having worked in WASH, and somelogical reasoning*

**Location Information**
The spatial distribution (regional, district level, section level) of columns/attributes.The column:`5420051|Location` contains information about the REGION, DISTRICT and CHIEFDOM and  in the form of `South|Bonthe|Bonthe Urban|Bonthe Town`. The data in this column was inconsistent, hence spatial join of data points and shapefiles of administrative boundaries.

*Retrieved region, district and chiefdom data*
